```python
import numpy as np
am = pd.DataFrame(np.zeros(shape=(3334,3334)), columns=df_new["diseaseId"].unique(), index=df_new["diseaseId"].unique())
```

```python
for dis in df_new['diseaseId'].unique():
    
    for gen in list(df_new[df_new['diseaseId'] == dis]['geneSymbol']):
        
        for gen_dis in df_new[df_new['geneSymbol'] == gen]['diseaseId']:          
        
            am.at[gen_dis,dis] += 1
```

```python
am
```

```python
nodes=pd.DataFrame(am.columns)
nodes.reset_index(inplace=True)
nodes.rename(columns={"index": "Id", 0 :"label"}, inplace=True)
nodes
```

```python
array=am.to_numpy()
```

```python
array
```

```python
N_art= array.diagonal()
```

```python
N_art
```

```python

#np.fill_diagonal(array, 0)
```

```python
array
```

```python
np.max(array)
```

```python
nd=list(nodes["Id"])
```

```python
Nt=df_new["geneSymbol"].nunique()
M_hyper =np.zeros((len(nd),len(nd)))
```

```python
N_art
```

```python
Nt
```

```python
import scipy.stats as s
for i in range(len(nd)):
  
  # number of articles tagged with tag i
  nA = N_art[i]
  
  for j in range(i+1,len(nd)):
  
    # number of articles tagged with tag j
    nB = N_art[j]
    
    # intersection
    k = array[i,j]
    
    # the first term is P(X>x), second term is P(X=x)

    p = s.hypergeom.sf(k, Nt, nA, nB) + s.hypergeom.pmf(k, Nt, nA, nB)

    
    M_hyper[i,j] = p
    M_hyper[j,i] = p
```

```python
M_hyper
```

```python
np.fill_diagonal(M_hyper, 1)
```

```python
M_hyper
```

```python
M_hyper[M_hyper==0] = min(M_hyper[M_hyper>0])
M_hyper_log = -np.log10(M_hyper)
M_hyper_log[np.array(M_hyper_log) < 0]=0
M_hyper_log
```

```python
n=M_hyper_log
n
```

```python
n[np.array(M_hyper_log) <2]=0
n
```

```python
n=np.tril(n, k=-1)
n
```

```python
len(nd)
```

```python
np.fill_diagonal(n, np.nan)
```

```python
M_Tr_hyper=pd.DataFrame(n,index=nd,columns=nd)
M_Tr_hyper
```
